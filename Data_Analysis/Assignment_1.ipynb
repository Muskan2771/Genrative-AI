### 1. Write code to create a 4x3 matrix with values ranging from 2 to 13.

### 2. Write code to replace the odd numbers by -1 in the array created by using random.randint( 1 , 50,20)

```python
arr2 = np.random.randint(1, 50, 20)
print("Original Array:", arr2)

arr2[arr2 % 2 != 0] = -1
print("After replacing odd numbers:", arr2)
```

### 3. Perform the following operations on an array of mobile phones prices 6999, 7500, 11999, 27899, 14999, 9999

```python
# a. Create a 1d-array of mobile phones prices
prices = np.array([6999, 7500, 11999, 27899, 14999, 9999])
print(prices)
```

```python
# b) Convert to float
prices = prices.astype(float)
print(prices)
```

```python
# c) Append new mobile price (13999)
prices = np.append(prices, 13999)
print(prices)
```

```python
# d) Reverse the array
prices = prices[::-1]
print(prices)
```

```python
# e) Apply 18% GST
prices = prices * 1.18
print(prices)
```

```python
# f) Sort prices in descending order
prices = np.sort(prices)[::-1]
print(prices)
```

```python
# g) Average mobile phone price
avg_price = np.mean(prices)
print("Average Price:", avg_price)
```

```python

# h) Difference between maximum & minimum price
price_diff = np.max(prices) - np.min(prices)
print("Difference:", price_diff)
```
